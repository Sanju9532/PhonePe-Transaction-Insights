```python
#importing the libraries
import pandas as pd
import os
import json

Agg_transaction_path = "C:/phonepay-project/data/aggregated/transaction/country/india/state/"
Aggregated_transaction=os.listdir(Agg_transaction_path)
Aggregated_transaction
columns_AT = {'State':[],'Year':[],'Quater':[],'Transaction_Type':[],'Total_no_transactions':[],'Transaction_Amount':[]}


for i in Aggregated_transaction:
    i_path = Agg_transaction_path + i + "/"
    transaction_list = os.listdir(i_path)
    
    for j in transaction_list:
        j_path = i_path  + j +"/"
        state_list = os.listdir(j_path)
        
        for k in state_list:
            k_path = j_path + k 
            Data = open(k_path,'r')
            datas = json.load(Data)
            for p in datas['data'] ['transactionData']:
                Name = p['name']
                Count = p['paymentInstruments'] [0] ['count']
                Amount = p['paymentInstruments'] [0] ['amount']
                columns_AT['State'].append(i)
                columns_AT['Year'].append(j)
                columns_AT['Quater'].append(int(k.strip('.json')))
                columns_AT['Transaction_Type'].append(Name)
                columns_AT['Total_no_transactions'].append(Count)
                columns_AT['Transaction_Amount'].append(Amount)


Agg_transaction = pd.DataFrame(columns_AT)

    
```
